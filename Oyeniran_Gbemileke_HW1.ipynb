From the scatter matrix above i can see some associations with the following predictors

1. Crim and age: as age increases crime rates (crim) increaases and vice versa, new buildings tends to have low crime rates
2. crim and dis: as dis increases crime rates (crim) decreases and vice versa
3. crim and medv: As medv increases crime rates (crim) decreases and vice versa, high values suburbs have less crime rates
4. Zn and indus: as zn increases the indus proportion decreases and vice versa
5. Zn and nox: as zn increases the porpotion of nox decreases and vice versa
6. dis and indus: as indus increases the proportion of dis decreases and vice versa
7. Age and lstat: as lstat increases the proportion of age also increases and vice versa, this means that more percentage of lower status people tends to stay at old houses.
8. lstat and medv: as the lstats increases the proportion or value of medv decreases and vice versa, this means that higer percentage of lower status tends to stay in suburbs with less median value of homes which really makes sense as they have less capital to afford the high value suburbs.
9. Medv and rm: as the medv increases the number or average room proportion increaases and vice versa, this means the higher the value of the median value of home in suburbs the more the rooms in the property, which makes sense as they spend more anount of money on the property and can afford additional rooms.

```python
# This code plots a 3 by 4 subplots for a scatter plot of crime rate against other predictors

fig, axes = subplots(nrows = 3, ncols = 4,
                    figsize = (21,21))
axes[0,0].scatter(boston["age"], boston["crim"]) # This is a code for the scatter plot of age against crim
axes[0,0].set_title("Scatter plot of age and crime rate")
axes[0,0].set_xlabel("Age")
axes[0,0].set_ylabel("Crim")

axes[0,1].scatter(boston["zn"], boston["crim"]) # This is a code for the scatter plot of zn against crim
axes[0,1].set_title("Scatter plot of zn and crime rate")
axes[0,1].set_xlabel("zn")
axes[0,1].set_ylabel("Crim")

axes[0,2].scatter(boston["indus"], boston["crim"]) # This is a code for the scatter plot of indus against crim
axes[0,2].set_title("Scatter plot of indus and crime rate")
axes[0,2].set_xlabel("Indus")
axes[0,2].set_ylabel("Crim")

axes[0,3].scatter(boston["chas"], boston["crim"]) # This is a code for the scatter plot of chas against crim
axes[0,3].set_title("Scatter plot of chas and crime rate")
axes[0,3].set_xlabel("Chas")
axes[0,3].set_ylabel("Crim")

axes[1,0].scatter(boston["nox"], boston["crim"], color = "teal") # This is a code for the scatter plot of nox against crim
axes[1,0].set_title("Scatter plot of nox and crime rate")
axes[1,0].set_xlabel("Nox")
axes[1,0].set_ylabel("Crim")

axes[1,1].scatter(boston["rm"], boston["crim"], color = "teal") # This is a code for the scatter plot of rm against crim
axes[1,1].set_title("Scatter plot of rm and crime rate")
axes[1,1].set_xlabel("Rm")
axes[1,1].set_ylabel("Crim")

axes[1,2].scatter(boston["dis"], boston["crim"], color = "teal") # This is a code for the scatter plot of dis against crim
axes[1,2].set_title("Scatter plot of dis and crime rate")
axes[1,2].set_xlabel("Dis")
axes[1,2].set_ylabel("Crim")

axes[1,3].scatter(boston["rad"], boston["crim"], color = "teal") # This is a code for the scatter plot of rad against crim
axes[1,3].set_title("Scatter plot of rad and crime rate")
axes[1,3].set_xlabel("Rad")
axes[1,3].set_ylabel("Crim")

axes[2,0].scatter(boston["tax"], boston["crim"], color = "grey") # This is a code for the scatter plot of age against crim
axes[2,0].set_title("Scatter plot of tax and crime rate")
axes[2,0].set_xlabel("Tax")
axes[2,0].set_ylabel("Crim")

axes[2,1].scatter(boston["ptratio"], boston["crim"], color = "grey") # This is a code for the scatter plot of ptratio against crim
axes[2,1].set_title("Scatter plot of ptratio and crime rate")
axes[2,1].set_xlabel("Pupil-Teacher ratio")
axes[2,1].set_ylabel("Crim")

axes[2,2].scatter(boston["lstat"], boston["crim"], color = "grey") # This is a code for the scatter plot of lstat against crim
axes[2,2].set_title("Scatter plot of lstat and crime rate")
axes[2,2].set_xlabel("Lstat")
axes[2,2].set_ylabel("Crim")

axes[2,3].scatter(boston["medv"], boston["crim"], color = "grey") # This is a code for the scatter plot of medv against crim
axes[2,3].set_title("Scatter plot of medv and crime rate")
axes[2,3].set_xlabel("Medv")
axes[2,3].set_ylabel("Crim")
```

From the scatter plots of crim against other predictors above it appears that the following predictors have some association with Crime rate (crim).

Age: as the proportion of age increases the crime rate (crim) also increases

Zn: as zn increase the crime rate (crim) decreases.

chas: it seems that suburbs without the charles river tends to have high crime rate (crim) while the ones with charles river has low crime rate (crim)

Dis: as the dis increases the crime rate (crim) decreases

Rad: ad the rad of the suburbs increases the crime rate increases

medv: as the median value of owner occupied homes increases the crime rate decreases. This really means that neighbourhood with high value i.e median value will have low crime rates which is ideal.
